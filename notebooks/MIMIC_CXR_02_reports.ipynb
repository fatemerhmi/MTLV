# Make all_cxr_study.csv

```python
def extract_findings(file_path):
    print(file_path)
    with open(file_path, 'r') as reader:
        report = reader.read()
    print("***"+ report + "***")
    x = re.search("(?<=FINDINGS:).*(?=IMPRESSION:)", report, flags = re.MULTILINE | re.DOTALL )
    findings = x.group().strip().replace('\n', '')
    print("###"+findings+"####")
    print("------------------------------------")
    return findings

root = "./data/MIMIC-III-CXR/mimic-cxr-reports/"
df_record_list['FINDINGS'] = df_record_list.apply(lambda row: \
                                                  extract_findings(root+row['path']), \
                                                  axis=1)
```
